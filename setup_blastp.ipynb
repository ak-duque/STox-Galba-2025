```python
# 5. Create a BLAST database
!makeblastdb -in "/content/ncbi/uniprot_sprot.fasta" \
-dbtype prot \
-out "/content/ncbi/uniprot_db"\
-title "uniprot_sprot_db" \
-parse_seqids
```

**PROTEOMICS**

Annotate Glycera peptide sets with BLASTp.

For each proteomic peptide dataset (P1, P2, S1, S2), we run 'blastp' against the local Swiss-Prot database we create.

P = proboscis; S = skin.

Duplicated samples.

Upload:

   'P1_Peptides.fasta' into /content/ncbi

   'P2_Peptides.fasta' into /content/ncbi

   'S1_Peptides.fasta' into /content/ncbi

   'S2_Peptides.fasta' into /content/ncbi

```python
# 6. Basic Local Alignment Search Tool for Proteins

# P1: Glycera proboscis sample

!blastp -query "/content/ncbi/P1_Peptides.fasta" \
        -db "/content/ncbi/uniprot_db" \
        -max_target_seqs 1 \
        -outfmt "10 delim=; qseqid sacc evalue pident qcovs stitle" \
        -evalue 1e-5 \
        -num_threads 10 \
        -out "/content/ncbi/BLASTp_Proteomica_P1_Peptides_annotation.csv"
```

```python
# 7. Basic Local Alignment Search Tool for Proteins

# P2: Glycera proboscis sample

!blastp -query "/content/ncbi/P2_Peptides.fasta" \
        -db "/content/ncbi/uniprot_db" \
        -max_target_seqs 1 \
        -outfmt "10 delim=; qseqid sacc evalue pident qcovs stitle" \
        -evalue 1e-5 \
        -num_threads 10 \
        -out "/content/ncbi/BLASTp_Proteomica_P2_Peptides_annotation.csv"
```

```python
# 8. Basic Local Alignment Search Tool for Proteins

# S1: Glycera skin sample

!blastp -query "/content/ncbi/S1_Peptides.fasta" \
        -db "/content/ncbi/uniprot_db" \
        -max_target_seqs 1 \
        -outfmt "10 delim=; qseqid sacc evalue pident qcovs stitle" \
        -evalue 1e-5 \
        -num_threads 10 \
        -out "/content/ncbi/BLASTp_Proteomica_S1_Peptides_annotation.csv"
```

```python
# 9. Basic Local Alignment Search Tool for Proteins

# S2: Glycera skin sample

!blastp -query "/content/ncbi/S2_Peptides.fasta" \
        -db "/content/ncbi/uniprot_db" \
        -max_target_seqs 1 \
        -outfmt "10 delim=; qseqid sacc evalue pident qcovs stitle" \
        -evalue 1e-5 \
        -num_threads 10 \
        -out "/content/ncbi/BLASTp_Proteomica_S2_Peptides_annotation.csv"
```
